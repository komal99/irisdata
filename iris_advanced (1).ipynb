```python
iris_data.to_csv('iris-data-clean.csv',index=False)
iris_data_clean=pd.read_csv('iris-data-clean.csv')
```

```python
import seaborn as sb
sb.pairplot(iris_data_clean,hue='class')
```

```python
all_inputs=iris_data_clean[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']].values
all_classes=iris_data_clean['class'].values
```

```python
all_inputs[:5]
```

```python
all_classes
```

```python
from sklearn.model_selection import train_test_split
(training_inputs,testing_inputs,training_classes,testing_classes)=train_test_split(all_inputs,all_classes,train_size=0.75,test_size=0.25,random_state=1)
```

```python
training_inputs
training_classes
```

```python
from sklearn.tree import DecisionTreeClassifier
#create the classifier
decision_tree_classifier=DecisionTreeClassifier()
#train the classifier on training set
decision_tree_classifier.fit(training_inputs,training_classes)
#validate the classifer
decision_tree_classifier.score(testing_inputs,testing_classes)
```

```python
prediction=decision_tree_classifier.predict(testing_inputs)
```

```python
probs=decision_tree_classifier.predict_proba(testing_inputs)
```
